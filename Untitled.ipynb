```python
from pystac import Catalog, Item, Link
from pystac.html import jinja_env

from datetime import datetime
```

```python
link1 = Link(rel="child", target="https://google.com", media_type="gif")
link2 = Link(rel="child", target="https://google.com/test", media_type="gif")
link3 = Link(rel="child", target="https://google.com/test2", media_type="gif")

item = Item("id",
            None,
            None,
            datetime.now(),
            properties={'entry1': 'some value', 'entry2': 1234},
            stac_extensions=["https://stac-extensions.github.io/datacube/v1.0.0/schema.json",
                            "https://stac-extensions.github.io/eo/v1.0.0/schema.json"])
item.add_link(link1)
item.add_link(link2)
item.add_link(link3)
```

```python
cat = Catalog("catId", "a catalog for some different things", "catalog title")
cat.add_item(item)
cat
```
